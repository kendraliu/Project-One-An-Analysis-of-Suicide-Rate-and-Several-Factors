```python
import pandas as pd
```

```python
suicide_rates_df = pd.read_csv('master.csv')
suicide_rates_df = suicide_rates_df.drop(['country-year',' gdp_for_year ($) ','generation'], axis=1)
suicide_rates_df = suicide_rates_df[suicide_rates_df['year'] == 2020]
suicide_rates_df
```

```python
suicide_rates_df = suicide_rates_df.dropna()
suicide_rates_df
```

```python
data2 = suicide_rates_df.groupby(["country"]).sum()
data2
```

```python
data1 = pd.DataFrame(data2["suicides_no"])
data1 = data1.reset_index()
data1
```

# Two ways to get population data

Option 1

```python
pop = pd.DataFrame(suicide_rates_df.groupby(["country"]).mean()["population"])
pop = pop.reset_index()
pop
```

Option 2

```python
a = suicide_rates_df["population"].unique()
```

```python
b = suicide_rates_df["country"].unique()
```

```python
pop2 = pd.DataFrame({"country": b, "population": a})
pop2
```

# Getting HDI data

```python
c = suicide_rates_df["HDI for year"].unique()
c
```

```python
data3 = suicide_rates_df.groupby(["country"]).mean()
data3 = pd.DataFrame(data3["HDI for year"])
data3 = data3.reset_index()
data3
```

# Calculating suicide rate

```python
calc = pd.merge(data1, pop, on="country")
calc
```

```python
calc["Suicide rate"] = calc["suicides_no"] / calc["population"]
calc = calc.sort_values(by="Suicide rate", ascending=False)
calc
```

# Merge all Data

```python
final = pd.merge(calc, data3, on="country")
final
```

```python
final.to_csv("SuicideRate_HDI.csv")
```
